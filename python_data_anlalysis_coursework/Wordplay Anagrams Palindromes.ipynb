```python
import math
import collections
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
%matplotlib inline
```

Load Dictionary

```python
words = [word.strip().lower() for word in open("words.txt","r").readlines()]
```

Create each possible combination of String

```python
words_sorted = {word:"".join(sorted(word)) for word in words}
def get_empty_list():
    return set()

letters_words = collections.defaultdict(get_empty_list)
for word,sorted_word in words_sorted.items():
    letters_words[sorted_word].add(word)
    
letters_words = {word:anagrams for word,anagrams in letters_words.items() if len(anagrams)>1}
```

Function to find anagrams of given word

```python
def find_anagrams(word):
    try:
        return letters_words["".join(sorted(word))]
    except KeyError:
        return set()


%time find_anagrams("dictionary")
%time find_anagrams("Siddhant")
```

Find all Palindromic pairs (All words that become another word when reversed)

```python
palindromes = collections.defaultdict(get_empty_list)
palindromes = {word: word[::-1] for word in words}
palindrome_pairs = {word: word[::-1] for word in words if word[::-1] in palindromes.keys()}
palindrome_pairs = {word:pair for word,pair in palindrome_pairs.items() if len(palindrome_pairs[word]) >= 1}
```
