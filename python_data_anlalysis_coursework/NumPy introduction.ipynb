Math with NUMPY

```python
arr = np.linspace(0,5 * math.pi, 64) #Create 64 equally spaced values between 0 and 5*pi
arr
```

```python
sinarr = np.sin(arr)
sinarr
```

```python
pp.figure(figsize=(16,9))
pp.plot(arr,sinarr,label = "sin")
pp.plot(arr, np.cos(arr), label = "cos")
pp.plot(arr, np.log(arr + 1), label = "log")

pp.legend()
```
